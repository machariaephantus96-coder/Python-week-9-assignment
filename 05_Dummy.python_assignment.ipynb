```python
# importing the libraries
```

```python
import requests
import pandas as pd
```

```python
# creating a variable for the url
```

```python
url = "https://dummyjson.com/users"
```

```python
# Getting a response from the URL
```

```python
data = requests.get(url)
data.status_code
```

```python
#getting the data in JSON from the URL and checking data type
```

```python
data_dummy = data.json()
```

```python
type(data_dummy)
```

```python
# extracting the values from the dictionary
# checking the data type
```

```python
customer_file = data_dummy["users"]
```

```python
type(customer_file)
```

```python
# normalizing the data
```

```python
customer_file = pd.json_normalize(customer_file)
```

```python
type(customer_file)
```

```python
# converting a JSON file into a relational data structure
```

```python
df = pd.DataFrame(customer_file)
df
```

```python
df.columns
```

```python
# converting the data into a CSV file
```

```python
df.to_csv("dummy_data.csv")
```
