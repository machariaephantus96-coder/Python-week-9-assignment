```python
# importing the libraries
```

```python
import requests
import pandas as pd
```

```python
# creating a variable for the url
```

```python
url = "https://dummyjson.com/carts"
```

```python
# Getting a response from the URL
```

```python
data = requests.get(url)
```

```python
#getting the data in JSON from the URL and checking data type
```

```python
data_carts = data.json()
type(data_carts)
```

```python
data_carts
```

```python
# Flattening products and keeping cart-level information
```

```python
df= pd.json_normalize(
    data_carts["carts"],
    record_path="products",
    meta=["id", "total", "discountedTotal", "userId", "totalProducts", "totalQuantity"],
     meta_prefix="cart_",
    errors="ignore")
df = df.rename(columns={"id": "product_id"})
    
```

```python

df
```

```python
df.to_csv("carts_endpoint.csv")
```
