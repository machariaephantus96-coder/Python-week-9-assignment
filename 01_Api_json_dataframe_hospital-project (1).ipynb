```python
# importing libraries
```

```python
import requests
import pandas as pd
```

```python
# creating a variable for the URL
```

```python
url = "https://raw.githubusercontent.com/MwangiJoseph-25/api-json-to-dataframe_Project/refs/heads/main/MOCK_DATA.json"
```

```python
# Getting a response from the URL
```

```python
data = requests.get(url)
data.status_code
```

```python
#getting the data in JSON from the URL and checking data type
```

```python
data_json = data.json()
```

```python
type(data_json)
```

```python
# converting a JSON file into a relational data structure
#slicing the rows to 100 
```

```python
df = pd.DataFrame(data_json[:100])
df
```
