```python
# importing libraries
```

```python
import requests
import pandas as pd
```

```python
# creating a variable for the url
```

```python
url = "https://dummyjson.com/products"
```

```python
# Getting a response from the URL
```

```python
data = requests.get(url)
data.status_code
```

```python
#getting the data in JSON from the URL and checking data type
```

```python
data_p = data.json()
type(data_p)
```

```python
# extracting the values from the dictionary
# checking the data type
```

```python
data_products = data_p["products"]
type(data_products)
```

```python
# normalizing the data
```

```python
products = pd.json_normalize(data_products)
```

```python
type(products)
```

```python
# converting a JSON file into a relational data structure
```

```python
df_products = pd.DataFrame(products[:10])
df_products
```

```python
# converting the data into a CSV file
```

```python
df_products.to_csv("products_data.csv")
```
